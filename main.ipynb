```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv("train.csv", index_col=['id'], parse_dates=['timestamp'])
```

```python
dataset.describe()
```

```python
i = 0;
for column in dataset:
    if i > 1:
        try:
            val = float(dataset[column].iloc[0])
        except:
            encoder = OneHotEncoder()
            label_encoder = LabelEncoder()
            data_label_encoded = label_encoder.fit_transform(dataset[column])
            dataset[column] = data_label_encoded
            data_feature_one_hot_encoded = encoder.fit_transform(dataset[[column]].as_matrix())
            
        dataset[column] = dataset[column].fillna(dataset[column].mean())
        
    i += 1
```

```python
dataset.to_csv("posttrain.csv")
```
